```python
import cv2
import glob
import pandas as pd
from imageio import imread,imsave
from skimage.transform import resize
from tqdm import tqdm
import dlib
import os
import random   
from mtcnn.mtcnn import MTCNN
import numpy as np
detector = MTCNN() 

input_shape = (240,240,3)

def random_id(length):
    number = '0123456789'
    alpha = 'abcdefghijklmnopqrstuvwxyz'
    id = ''
    for i in range(0,length,2):
        id += random.choice(number)
        id += random.choice(alpha)
    return id

train_paths = glob.glob("images/*")
savepic="./pic/"+random_id(20)+".jpg"
print(train_paths)

df_train = pd.DataFrame(columns=['image', 'label', 'name'])
# Initial Parameter
a=0
size=4   
# Load image
for i,train_path in tqdm(enumerate(train_paths)):
    name = train_path.split("\\")[-1]
    images = glob.glob(train_path + "/*")
    for image in images:
        df_train.loc[len(df_train)]=[image,i,name]        
print(df_train)
# Begin Process
for img_path in df_train.image:
    print(img_path)
    image = imread(img_path)
    mini = cv2.resize(image, (image.shape[1] // size, image.shape[0] // size))
    faces = detector.detect_faces(mini)
    for f in faces: 
        x, y, w, h = [v * size for v in f['box']]
        face_img = image[y:y + h, x:x + w]

        ##Xu ly duong dan:
        temp=img_path
        temp=temp.split("\\")
        temp=temp[2]        
        # # ket thuc
        imsave("./pic/"+temp,face_img)
        #Xu li ten 
        name="./pic/"+random_id(40)+'.jpg'
        os.rename("./pic/"+temp,name)
```
