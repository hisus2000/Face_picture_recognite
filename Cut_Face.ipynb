```python
import cv2
import glob
import pandas as pd
from imageio import imread,imsave
from skimage.transform import resize
from tqdm import tqdm
import dlib
import os
train_paths = glob.glob("images/*")
print(train_paths)
df_train = pd.DataFrame(columns=['image', 'label', 'name'])
# Load image
for i,train_path in tqdm(enumerate(train_paths)):
    name = train_path.split("\\")[-1]
    images = glob.glob(train_path + "/*")
    for image in images:
        df_train.loc[len(df_train)]=[image,i,name]        
print(df_train)
#Create folder
try:
    name_folder=df_train["name"].unique()
    for i in name_folder:
        name=str(i)
        dic="./process/" +name
        os.mkdir(dic) 
except:
    pass
# Begin Process
try:    
    for img_path in df_train.image:
        print(img_path)
        image = imread(img_path)
        hogFaceDetector = dlib.get_frontal_face_detector()
        faceRects = hogFaceDetector(image, 0)
        faceRect = faceRects[0]
        if (faceRect==None):
            continue

        x1 = faceRect.left()
        y1 = faceRect.top()
        x2 = faceRect.right()
        y2 = faceRect.bottom()

        face = image[y1:y2,x1:x2]
        temp=img_path.split("\\")      
        new_dic="./process"+"/"+temp[1]+"/"+temp[2]
        imsave(new_dic,face)
except:
    pass 

data_save = glob.glob("process/*")
data_train = pd.DataFrame(columns=['image', 'label', 'name'])
# Load image
for i,data_save in tqdm(enumerate(data_save)):
    name = data_save.split("\\")[-1]
    images = glob.glob(data_save + "/*")
    for image in images:
        data_train.loc[len(data_train)]=[image,i,name]      
with open('test.txt', 'w') as f:
    for img_path in data_train.image:
        temp=img_path.split("\\")
        name="./"+temp[1]+"/"+temp[2]+"\t"+temp[1]
        f.write(name)
        f.write('\n')      
```
